```python
%matplotlib inline
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pylab as plt
plt.rcParams.update({'font.size': 16})
import matplotlib.colors as mcolors
import glob,os
import datetime as dt
```

```python
Pow=pd.read_csv('Powell4qubits.txt',sep="\t")
bfgs=pd.read_csv('BFGS4qubits.txt',sep="\t")
cg=pd.read_csv('CG4qubits.txt',sep="\t")
cobyla=pd.read_csv('COBYLA4qubits.txt',sep="\t")
lbfgs=pd.read_csv('L-BFGS-B4qubits.txt',sep="\t")
slsqp=pd.read_csv('SLSQP4qubits.txt',sep="\t")
trustc=pd.read_csv('trust-constr4qubits.txt',sep="\t")
```

```python
fig,ax=plt.subplots(2,1,figsize=(20,25))
Depth=np.arange(1,11)

sns.lineplot(x=Depth, y=Pow.accuracy,ax=ax[0], label="Powell", marker='o', markersize=9, color="deepskyblue")
sns.lineplot(x=Depth, y=Pow.time,ax=ax[1], label="Powell", marker='o', markersize=9, color="deepskyblue")

sns.lineplot(x=Depth, y=bfgs.accuracy,ax=ax[0], label="BFGS", marker='o', markersize=9, color="deeppink")
sns.lineplot(x=Depth, y=bfgs.time,ax=ax[1], label="BFGS", marker='o', markersize=9, color="deeppink")

sns.lineplot(x=Depth, y=cg.accuracy,ax=ax[0], label='CG', marker='o', markersize=9, color="mediumblue")
sns.lineplot(x=Depth, y=cg.time,ax=ax[1], label='CG', marker='o', markersize=9, color="mediumblue")

sns.lineplot(x=Depth, y=cobyla.accuracy,ax=ax[0], label='COBYLA', marker='o', markersize=9, color="blueviolet")
sns.lineplot(x=Depth, y=cobyla.time,ax=ax[1], label='COBYLA', marker='o', markersize=9, color="blueviolet")

sns.lineplot(x=Depth, y=lbfgs.accuracy,ax=ax[0], label='L-BFGS-B', marker='o', markersize=9, color="limegreen")
sns.lineplot(x=Depth, y=lbfgs.time,ax=ax[1], label='L-BFGS-B', marker='o', markersize=9, color="limegreen")

sns.lineplot(x=Depth, y=slsqp.accuracy,ax=ax[0], label='SLSQP', marker='o', markersize=9, color="darkorange")
sns.lineplot(x=Depth, y=slsqp.time,ax=ax[1], label='SLSQP', marker='o', markersize=9, color="darkorange")

sns.lineplot(x=Depth, y=trustc.accuracy,ax=ax[0], label='Trust-constr', marker='o', markersize=9, color="brown")
sns.lineplot(x=Depth, y=trustc.time,ax=ax[1], label='Trust-constr', marker='o', markersize=9, color="brown")

ax[0].grid(True)
ax[1].grid(True)
ax[0].set_ylabel(r"$\log(1/\epsilon)$")
ax[0].set_xlabel(r"Depth")
ax[1].set_ylabel("Time [s]")
ax[1].set_xlabel("Depth")
ax[0].set_title("4 qubits simulation")
```
